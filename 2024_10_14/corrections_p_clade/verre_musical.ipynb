```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np

sample_rate, data = wavfile.read('data/SingingGlass.wav')
N = len(data)
```

```python
T = N/sample_rate
delta_T = 1/sample_rate
```

```python
liste_t = np.arange(N)*delta_T

plt.plot(liste_t, data)
```

```python
plt.plot(liste_t[10000:10100], data[10000:10100])
plt.xlabel('Temps [s]')
```

```python
data_tilde = np.fft.rfft(data)
freqs = np.fft.rfftfreq(len(data), d=1/sample_rate)

mask = (freqs>1600) & (freqs<1620)
plt.semilogy(freqs[mask], np.abs(data_tilde)[mask])
plt.grid()
```
