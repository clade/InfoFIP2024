```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np

sample_rate, data = wavfile.read('data/machine_a_laver.wav')
N = len(data)
```

```python
liste_t = np.arange(N)/sample_rate
plt.plot(liste_t, data)
```

```python
data_tilde = np.fft.rfft(data)
freqs = np.fft.rfftfreq(len(data), d=1/sample_rate)

plt.loglog(freqs, np.abs(data_tilde))
plt.ylim(1E-2, None)
```

```python
mask = (freqs>10) & (freqs<50)
plt.loglog(freqs[mask], np.abs(data_tilde)[mask])
plt.ylim(1E-2, None)

i_max = np.argmax(np.abs(data_tilde)[mask])
freq_max = freqs[mask][i_max]
amp_max = np.abs(data_tilde)[mask][i_max]

plt.loglog(freq_max, amp_max, 'o')
plt.ylim(1, None)

plt.text(freq_max, amp_max,f' {freq_max:.2f} Hz')
plt.xlabel('Frequence [Hz]')
```
