```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussienne(L, sigma, N):
    Delta_L = L/N
    x = np.arange(N)*Delta_L - L/2
    return x, np.exp(-x**2/(2*sigma**2))

x, y = gaussienne(5, .5, N=1000)

plt.plot(x, y)
```

```python
L = 25
sample_rate = 40
N = sample_rate * L
sigma = 1.5

x, y = gaussienne(L, sigma, N)

y_tilde = np.fft.fft(y)
freqs = np.fft.fftfreq(N, L/N)

plt.plot(np.fft.fftshift(freqs), np.abs(np.fft.fftshift(y_tilde)*L/N))

omegas = 2*np.pi*freqs
y_tilde_theo = np.sqrt(2*np.pi)*sigma*np.exp(-omegas**2*sigma**2/2)

plt.plot(np.fft.fftshift(freqs), np.fft.fftshift(y_tilde_theo))
```

```python
plt.plot(np.fft.fftshift(freqs), np.fft.fftshift(y_tilde))
plt.xlim(-3, 3)
```
