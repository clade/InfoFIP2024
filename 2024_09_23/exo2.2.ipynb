```python
import numpy as np
```

```python
M = 1000

nuage = 2*np.random.rand(M, 2) - 1

```

```python
import matplotlib.pyplot as plt

plt.plot(nuage[:,0], nuage[:,1], 'o')

R = 1
mask = nuage[:,0]**2 + nuage[:,1]**2 < R**2

plt.plot(nuage[mask,0], nuage[mask,1], 'o')
plt.axis('equal')
```

```python
M = 100_000_000

nuage = 2*np.random.rand(M, 2) - 1
mask = (nuage[:,0]**2 + nuage[:,1]**2) < R**2
```
