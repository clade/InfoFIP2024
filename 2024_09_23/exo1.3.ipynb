```python
with open("cac40.csv") as f:
    data = f.readlines()
print(data[0])
data = data[1:]
    
for line in data[:10]:
    print(line.strip())
```

```python
plus_haut = []
for line in data:
    plus_haut.append(float(line.strip().split(',')[2]))
    
max(plus_haut)
```

```python
def date(annee, mois, jour):
    return f"{annee}-{mois:02d}-{jour:02d}"

date(2024, 9, 23)
date(1515, 9, 13)
```

```python
def plus_bas(annee, mois, jour):
    date_str = date(annee, mois, jour)
    for line in data:
        split_line = line.strip().split(',')
        if split_line[0]==date_str:
            return float(split_line[3])
    print(f"La date {date_str} n'existe pas")

plus_bas(2024, 8, 4)
```

```python
cac_40 = {}
for line in data:
    Date, Open, High, Low, Close, Adj_Close, Volume = line.strip().split(',')
    cac_40[Date] = {
        'open':float(Open),
        'high':float(High),
        'low':float(Low),
        'close':float(Close),
        'volume':float(Volume)        
    }
    
cac_40['2020-10-05']['high']
```

```python
import json
with open('cac40.json', 'w') as f:
    json.dump(cac_40, f)
```

```python
with open('cac40.json') as f:
    print(json.load(f))
```

```python
from dateutil.parser import parse

X = []
Y = []
for key, val in cac_40.items():
    X.append(parse(key))
    Y.append(val['high'])
    
import matplotlib.pyplot as plt

plt.plot(X, Y)
             
```
