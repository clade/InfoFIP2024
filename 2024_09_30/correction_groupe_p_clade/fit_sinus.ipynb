```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x_data, y_data = np.loadtxt('fit_sinus.dat', unpack=True)
```

```python
plt.plot(x_data, y_data, 'o')
plt.xlabel('Frequence [Hz]')
plt.ylabel('Amplitude')
plt.grid()
```

```python
def frange(x, centre, interfrange, moyenne, amplitude):
    return moyenne + amplitude * np.cos(2*np.pi*(x-centre)/interfrange)

x = np.linspace(-220, 220)

plt.plot(x_data, y_data, 'o')
plt.xlabel('Frequence [Hz]')
plt.ylabel('Amplitude')
plt.grid()

p_ini = [0, 200, 0.2, 0.1]
#plt.plot(x, frange(x, *p_ini))

popt, pcov = curve_fit(frange, x_data, y_data, p_ini)
plt.plot(x, frange(x, *popt))
```

```python
x_opt = popt[0]
sigma_x = np.sqrt(pcov[0, 0])

print(x_opt)
print(sigma_x)
```
